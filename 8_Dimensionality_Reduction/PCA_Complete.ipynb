# Principal Component Analysis (PCA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
wine_data = pd.read_csv('Wine_data.csv')
```

Atributes for this dataset
    
    1) Alcohol 
    2) Malic acid 
    3) Ash 
    4) Alcalinity of ash 
    5) Magnesium 
    6) Total phenols 
    7) Flavanoids 
    8) Nonflavanoid phenols 
    9) Proanthocyanins 
    10)Color intensity 
    11)Hue 
    12)OD280/OD315 of diluted wines 
    13)Proline 

```python
wine_data.head()
```

### Preprocessing step

    We have to scale our data before using our PCA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(wine_data.values)
print(scaled_data)
```

```python
print(scaled_data.shape)
```

As we can see we have 13 features in this data, and it cannot be visualised. We will need to reduce it with PCA to 2 components.


### Sklearn PCA

```python
from sklearn.decomposition import PCA
```

```python
sklearn_pca = PCA(n_components=2)
```

```python
reduced_data = sklearn_pca.fit_transform(wine_data.values)
print(reduced_data)
print (sklearn_pca.explained_variance_ratio_)
```

```python
print(wine_data.shape)
print(reduced_data.shape)
```

```python
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], label='Wine points reduced')
plt.title("PCA sklearn")
plt.legend()
plt.show()
```

```python


reduced_data = sklearn_pca.fit_transform(wine_data.values)
print(reduced_data)
print (sklearn_pca.explained_variance_ratio_)

#pca = PCA(n_components=2)
#principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = reduced_data
             , columns = ['principal component 1', 'principal component 2'])
principalDf.head(20)

finalDf = pd.concat([principalDf, wine_data[['Customer_Segment']]], axis = 1)
finalDf.head()
```

```python



fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [1, 2, 3]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Customer_Segment'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
