```python
from matplotlib import pyplot as plt 
#1. Line Graph
x=[5,7,10,15,25,30]
y=[6,10,15,20,30,33]

plt.figure(figsize=(4,3)) # width & height are the figure in inches.
plt.plot(x,y,marker='o',color='red')
plt.xlabel('Sales')
plt.ylabel('Marketing Exp')
plt.title('Same Line Graph')
plt.show()
```

```python
#2. Multi-Line Graph
x=[5,7,10,15,25,30]
y=[6,10,15,20,30,33]

x2=[9,10,15,25,30]
y2=[6,12,17,25,30]

plt.figure(figsize=(4,2)) # width & height are the figure in inches.
plt.plot(x,y,marker='o',color='red')
plt.plot(x2,y2,marker='*',color='green')

plt.xlabel('Sales')
plt.ylabel('Marketing Exp')
plt.title('Same Line Graph')
plt.show()
```

```python
#3. Bar Chart
langs=['C','C++','Java','Python','Php']
student=[23,45,21,70,12]

plt.figure(figsize=(7,5))
plt.bar(langs,student,color='blue',width=0.7)

plt.xlabel('Courses')
plt.ylabel('Student Enrolled')
plt.title('Course Enrollments')

plt.show()
```

```python
#3. Horizontal Bar Chart
langs=['C','C++','Java','Python','Php']
student=[23,45,21,70,12]

plt.figure(figsize=(7,5))
plt.barh(langs,student,color='#19e0cc') #'blue') 

plt.ylabel('Courses')
plt.xlabel('Student Enrolled')
plt.title('Course Enrollments')

plt.show()
```

```python
#Python Matplotlib: Bar Graph

plt.figure(figsize=(4,3))
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="BMW",width=.5,color="lightgreen")
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="Merc",width=.5,color="Green")

plt.legend()
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')

plt.show()
```
