```python
# Selections: Specific Columns or Rows 
# Label based selection: at[], loc[]
# Index based selection: iloc[] 
# Conditional Selections: df[conditions(s)]


# .loc	Label-based	Access rows and columns by labels	Selecting by labels or conditions
# .iloc	Integer position	Access rows and columns by positions	Selecting by row and column index
# .at	Label-based (scalar)	Fast access to a single value by labels	Single element access or modification

```

```python
# Selections: Specific Columns or Rows 
# Label based selection: at[], loc[]
df = pd.read_csv('diabetes.csv',index_col=['Pid'])
print(df)
# locate & updation --> .at[]
print("Age of ID:10004", df.at[1004,'Age'])
df.at[1004,'Age']=35
print("Age of ID:10004", df.at[1004,'Age'])
```

```python
# df.loc[row_index range, list of columns]
print(df.loc[1001:1005])
```

```python
# df.loc[row_index range, list of columns]

print(df.loc[  [1001,1005],['Pregnancies','BloodPressure','Age']])

df_new = df.loc[  [1001,1002, 1003, 1004, 1005],['Pregnancies','BloodPressure','Age']]
df_new 
```

```python
# Index based .iloc[] 
# can not used custom indices 
# df.iloc[row index, column index]
print(df.iloc[0:5,1:4])

print(df.iloc[[1,5],[1,4]])
```

```python
#Adding and Dropping Columns 
# .iloc 
# Adding a new column 
df.loc[:, ['City']]='Mumbai'
df.loc[:, ['Outcome']]=1
print(df)
```

```python
# Droping/Delete Columns and Rows
# only 1 one function available --> drop function column as well row
df.drop(['Outcome'],axis=1) # axis=1 --> columns  
# Pandas protect the deletion of the original data
# axis=0 --> for rows 
print(df)

# create a new structure
df_updated = df.drop(['Outcome'],axis=1) 
print(df_updated)

# Alter the original structure data frame 
df.drop(['Outcome'],axis=1, inplace=True) 

# Deleting Rows
df.drop([1001],axis=0)   # axis=0 --> for rows 

df.sort_values(by='Age', ascending=False)
```
