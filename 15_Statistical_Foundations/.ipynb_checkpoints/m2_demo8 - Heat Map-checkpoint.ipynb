<h2>Demo 8: Visualize Data using Heat Map</h2>

In this demo, you will be shown how heatmaps help visualize collinearity of multivariate data.

```python
#Step1: Import the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


# Step 2: Read the data from the uploaded CSV file
df = pd.read_csv('cars.csv')

# Step 3: Fetch the preview of the dataset
print(df.head())

# Step 4: Prepare the dataset for the heatmap
# Drop non-numeric columns
df_numeric = df.select_dtypes(include=[float, int])

# Step 5: Create a heatmap
fig, ax = plt.subplots(figsize=(10, 6))  # Set the figure size
sns.heatmap(df_numeric.corr(), center=0, cmap='Reds', annot=True)  # Add annot=True for correlation values
ax.set_title('Multi-Collinearity of Car Attributes')
plt.show()
```

This heatmap represents the collinearity of the multiple variables in the dataset. 
Since the model names are not numerical values, the 'unnamed' column was dropped off.

Dark red represents a positive correlation, while light white is a negative correlation.

```python
# Step 3: Drop non-numeric columns
#df_numeric = df.select_dtypes(include=[float, int])  # Select only numeric columns

# Step 5: Annotate the heatmap with numeric values
fig, ax = plt.subplots(figsize=(10, 6))  # Set figure size
sns.heatmap(df_numeric.corr(), center=0, cmap='BrBG', annot=True)  # Add annotations
ax.set_title('Correlation Heatmap of Numeric Attributes')
plt.show()
```

You can see that <b>disp</b> and <b>cylinder</b> are highly correlated (Features with high correlation are more <b>linearly dependent</b> and hence have almost the same effect on the dependent variable <b>mpg</b>,  i.e., 0.85 and 0.85).

When two features have high correlation, we can drop one of the two features and still produce a fairly accurate prediction model.
