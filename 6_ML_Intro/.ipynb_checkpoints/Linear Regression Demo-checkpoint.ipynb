```python
#Problem Statement: Predict sales of a product based on TV, Radio and newspaper adds
```

```python
#Step 01
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#For Evaluation 
import sklearn.metrics
```

```python
#Step 02 
#Reading data
data=pd.read_csv("Advertising.csv")
data.head() #Top 5 rows
```

```python
data.shape
```

```python
#EDA-Know more about data
data.info()
print("***")
```

```python
#Checking Null Values
data.isnull().sum()
```

```python
data.describe()
```

```python
#Checking Linear Relationship
plt.scatter(data['TV'],data['sales'])
plt.xlabel("TV Add exp")
plt.ylabel("Sales")
plt.show()
```

```python
#Step: 3 Segrate Independent and Dependent Variable
x=data['TV'].values.reshape(-1,1)   #Independent
y=data['sales'].values.reshape(-1,1) #Dependent
print(type(x))
print(x)
```

```python
#Step: 04 Train The model
lin_reg=LinearRegression()
lin_reg.fit(x,y)  #Training Model
```

```python
#Step:05 Predict values
y_pred=lin_reg.predict(x)
print(y_pred)
```

```python
#Coeficients of Lin Reg
print("B1:",lin_reg.coef_)
print("B0:",lin_reg.intercept_)
```

```python
# Sales=7.032+0.0475*TV   Equation Lin reg in this case
```

```python
plt.scatter(data['TV'],data['sales'])
plt.plot(data['TV'],y_pred,color='red')
plt.xlabel("TV Add exp")
plt.ylabel("Sales")
plt.show()
```

```python
#Step:06 Evaluate the model
from sklearn import metrics
import math
print("MSE:",metrics.mean_squared_error(y_pred,y))
print("RMSE:",math.sqrt(metrics.mean_squared_error(y_pred,y)))
print("MAE:",metrics.mean_absolute_error(y_pred,y))
```

```python
#Checking R2
print("Accuracy(R2):",lin_reg.score(x,y))
```

```python
#Predictions on Unknow inputs
TV_add=int(input("Enter TV add Exp:"))
print("Estimated Sales:",lin_reg.predict([[TV_add*1000]]))

# Sales=7.032+0.0475*TV   Equation Lin reg in this case
```

```python
#Multiple Linear Rgeression

#Step: 3 Segrate Independent and Dependent Variable
x=data[['TV','radio','newspaper']].values  #Independent
y=data['sales'].values.reshape(-1,1) #Dependent
print(type(x))
print(x)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x,y)  #Training Model
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```
