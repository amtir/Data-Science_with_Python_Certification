```python
# Linear Regression 

# Problem statement: Predict sales of a product based on TV, Radio, and Newspaper Adds.
```

```python
# Step1 import libraries 

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
# For Evaluation 
import sklearn.metrics 
```

```python
#Step2: reading data 
data = pd.read_csv('Advertising.csv')
print("shape: ", data.shape)
# EDA 
print("EDA")
print(data.info())
print(data.describe())
print(data.head())
```

```python
# check for null values 
print(data.isnull())
data.isnull().sum()
```

```python
# Step3: Segregate Independent and ependent variables 
sales=data.sales
tv=data.TV
print("type", type(tv))
radio=data.radio
newspaper=data.newspaper
tv.head()
x=data.TV.values.reshape(-1,1)
y=data.sales.values.reshape(-1,1)
```

```python
# Check relationship 
plt.scatter(tv, sales)
plt.xlabel("TV Add exp")
plt.ylabel("Sales")
plt.title("sales=F(tv)")
plt.show()
```

```python
# Step 4: Train the Model 
lin_reg = LinearRegression() # create the model empty default template 
lin_reg.fit(x, y) # Training the model, Train the supervised model 

```

```python
# Step 5: Predict values 
y_pred = lin_reg.predict(x)
#print(y_pred)
```

```python
#Error
err = y - y_pred
#print(err)
```

```python
# Coefficients of Linear Regression
print("B1: ", lin_reg.coef_)
print("B0: ", lin_reg.intercept_)

```

```python
# Sales = 7.03259355 + 0.04753664*TV Equation linear regression 
plt.scatter(tv, sales)
plt.xlabel("TV Add exp")
plt.ylabel("Sales")
plt.title("sales=F(tv)")
plt.plot(tv, y_pred, color="red")
plt.show()
```

```python
# Step 6:  Evaluate the model 
import math
print("Mean Squared Error MSE: ", sklearn.metrics.mean_squared_error(y_pred, y))
print("Root Mean Squared Error RMSE: ", math.sqrt(sklearn.metrics.mean_squared_error(y_pred, y)))
print("Mean Absolute Error MAE: ", sklearn.metrics.mean_absolute_error(y_pred, y))
# Checking R2
print("Accuracy R2: ", lin_reg.score(x, y))
print("Accuracy R2: ", lin_reg.score(x, y_pred))
```

```python
# Predictions on Unknown inputs
TV_add = int(input("Enter TV add Exp: "))
print("Estimated sales: ", lin_reg.predict([[TV_add*1000
                                         ]]))
```
