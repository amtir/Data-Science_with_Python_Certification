```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
adv = pd.read_csv('Advertising.csv')

print(adv.info())
print(adv.describe())
print(adv.head())

# ------------------------------
# Part 1: Simple Linear Regression (TV ads vs Sales)
# ------------------------------

# Selecting the independent (X) and dependent (y) variables
X_simple = adv[['TV']]
y = adv['sales']

# Initialize and train the model
lin_reg_simple = LinearRegression()
lin_reg_simple.fit(X_simple, y)

# Coefficients B0 (intercept) and B1 (slope) for TV ads
B0_simple = lin_reg_simple.intercept_
B1_simple = lin_reg_simple.coef_[0]

print("Simple Linear Regression:")
print(f"B0 (Intercept): {B0_simple}")
print(f"B1 (Coefficient for TV): {B1_simple}")

# Predicting sales based on TV ad spend
y_pred_simple = lin_reg_simple.predict(X_simple)

# Plotting the simple linear regression line
plt.scatter(adv['TV'], adv['sales'], color='blue', label='Actual Sales')
plt.plot(adv['TV'], y_pred_simple, color='red', label='Regression Line')
plt.xlabel("TV Ads Spend")
plt.ylabel("Sales")
plt.title("Simple Linear Regression: TV Ads vs Sales")
plt.legend()
plt.show()

# Equation
print(f"Sales = {B0_simple:.3f} + {B1_simple:.4f} * TV")

# ------------------------------
# Part 2: Multiple Linear Regression (TV, Radio, Newspaper vs Sales)
# ------------------------------

# Selecting the independent (X) variables and dependent (y) variable
X_multi = adv[['TV', 'radio', 'newspaper']]

# Initialize and train the model
lin_reg_multi = LinearRegression()
lin_reg_multi.fit(X_multi, y)

# Coefficients B0 (intercept), B1, B2, B3 for TV, radio, and newspaper ads
B0_multi = lin_reg_multi.intercept_
B1_multi, B2_multi, B3_multi = lin_reg_multi.coef_

print("Multiple Linear Regression:")
print(f"B0 (Intercept): {B0_multi}")
print(f"B1 (Coefficient for TV): {B1_multi}")
print(f"B2 (Coefficient for Radio): {B2_multi}")
print(f"B3 (Coefficient for Newspaper): {B3_multi}")

# Predicting sales based on all three ad spends
y_pred_multi = lin_reg_multi.predict(X_multi)

# Equation for multiple linear regression
print(f"Sales = {B0_multi:.3f} + {B1_multi:.4f} * TV + {B2_multi:.4f} * Radio + {B3_multi:.4f} * Newspaper")

# Optional: Model evaluation for both models
mse_simple = mean_squared_error(y, y_pred_simple)
r2_simple = r2_score(y, y_pred_simple)
mse_multi = mean_squared_error(y, y_pred_multi)
r2_multi = r2_score(y, y_pred_multi)

print("Evaluation:")
print("Simple Linear Regression MSE:", mse_simple, "R^2:", r2_simple)
print("Multiple Linear Regression MSE:", mse_multi, "R^2:", r2_multi)
```
